```python
## https://techdevguide.withgoogle.com/resources/former-coding-interview-question-word-squares/#!
#https://www.programiz.com/python-programming/online-compiler/?ref=6e314d24  
#complete where you find xyz

import numpy as np

#recursion steps: 
#                  end condition with simplest task (when we arrive to the basic case ?)
#                  execute n-th operation for semplification


def f(used, lenInputList, data):
    
    if (len(used) == lenInputList):
        print("\n\n# --------------Current data-------------- #");
        for j in used:
            print(data[j]);
            # xyz append data[j] to the tmp matrix
            #print(len(data[0]));
        print("# ---------------------------------- #\n\n");
        
        # check the word square using data == [[...], [...], [...]]             xyz
    
    for i in range(lenInputList):
        if (i not in used):
            used.append(i);
            f(used, lenInputList, data);
            used.remove(i);
    return;





#inputData = np.array([["B", "A", "L", "L"],
#             ["A","R","E","A"]]);
inputData = np.array([["B", "A", "L", "L"],
                      ["A", "R", "E", "A"],
                      ["L", "E", "A", "D"],
                      ["L", "A", "D", "Y"]]);

checkPassed = True;
indRow = 0;
firstLen = -1;
for tmpRow in inputData:
    #for tmpElem in tmpRow:
    if (firstLen == -1):
        firstLen = len(tmpRow);
        #print(tmpRow)
        print("First len found: ", firstLen);
    elif (len(tmpRow) != firstLen):
        print("NO WORD SQUARE. First check not passed !");
        checkPassed = False;
        break;
    
    # check on the equality between inputData[indRow] and inputData[0:len(inputData), indRow]
    # indRow
    # indCol
    indIncr = 0;
    verticalEl = list();
    while (indIncr < len(inputData)):
        verticalEl.append(inputData[indIncr, indRow]);
        indIncr += 1;
        #
    print(verticalEl, "  ", inputData[indRow])
    if (len(verticalEl) == len(inputData[indRow])):
        comparison = verticalEl == inputData[indRow]
        if (comparison.all() == False):
            print("Second check not passed ! ", inputData[indRow], " not equal to ", verticalEl);
            checkPassed = False;
            break;
    else:
        print("Third check not passed ! Len mismatch ");
        checkPassed = False;
        break;
    
    indRow += 1;
    
if (checkPassed == False):
    print("NOT WORD SQUARE");
else:
    print("YES WORD SQUARE");
    
    
    
    
    
    
    
    
# Second Task
checkPassed_ = True;
secondInputList = ["AB", "BC", "CD"]; #["AREA", "BALL", "DEAR", "LADY", "LEAD", "YARD"]
#check if all the words have the same line
generalLen = -1;
for tmp in secondInputList:
    if (generalLen == -1):
        generalLen = len(tmp);
    elif (len(tmp) != generalLen):
        checkPassed_ = False
        break;

print("\n\nSecond Task");
if (checkPassed_ == True):
    #print("\nPreliminary check passed");
    # generate the matrix   len(secondInputList, generalLen)
    inputData_ = np.empty((len(secondInputList), generalLen), dtype=str);
    #print(inputData_);
    index = 0
    for tmp in secondInputList:
        internalIndex = 0;
        for tmpChar in tmp:
            inputData_[index][internalIndex] = tmpChar;
            internalIndex += 1;
        index += 1;
    print("Generated ", inputData_);
    
    #start the generation
    f(list(), len(secondInputList), inputData_);
    
else:
    print("\nPreliminary check not passed");
```
